As per the data there are no wrong values entered in the data.

```python
# First we drop the na values in the horsepower column
df['horsepower'] = df['horsepower'].dropna()
```

```python
df
```

```python
df['horsepower'] = df['horsepower'].astype('int32')
```

```python
df['name'] = df['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

```python
df['name'].unique()
```

Now, Here we can create a new column called 'Company' from the name column. we will consider the first word of the name column.

```python
df['company'] = df['name'].str.split(' ').str.get(0)
```

```python
df
```

```python
df['company'].unique()
```

Now if you check the company column you can see that there are lot of wrong values entered in the column.
Precisely,

    vw = Volkswagon

    hi 1200d = ih (international harvester)

    vokswagen = Volkswagon

    chevroelt = chevrolet

    toyouta = toyota

    Maxda = mazda

    chevy = chevrolet

```python
# To tackle this problem we cretae a dictionary
dict = {'vw':'volkswagen','hi':'ih','vokswagen':'volkswagen','chevroelt':'chevrolet','toyouta':'toyota','maxda':'mazda','chevy':'chevrolet'}
```

```python
df['company'] = df['company'].replace(dict)
```
